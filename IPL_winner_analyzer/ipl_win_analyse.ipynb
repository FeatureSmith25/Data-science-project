```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('matches.csv')
df.head(10)
```

```python
df.isnull().sum()
```

```python
df['city']=df['city'].fillna(df['city'].mode()[0],inplace=True)
df['player_of_match']=df['player_of_match'].fillna(df['player_of_match'].mode()[0],inplace=True)
df['winner']=df['winner'].fillna(df['winner'].mode()[0],inplace=True)
df['umpire1']=df['umpire1'].fillna(df['umpire1'].mode()[0],inplace=True)
df['umpire2']=df['umpire2'].fillna(df['umpire2'].mode()[0],inplace=True)
df['umpire3']=df['umpire3'].fillna(df['umpire3'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.countplot(x='winner',data=df,palette='Set2',hue='winner')
plt.xticks(rotation=45, ha='right')
```

```python
team1_count=df['team1'].value_counts()
team2_count=df['team2'].value_counts()
total_matches=team1_count.add(team2_count, fill_value=0)
```

```python
total_wins=df['winner'].value_counts()
total_wins
```

```python
team_mapping = {
    'Delhi Daredevils': 'Delhi Capitals',       # Same team, naam badla
    'Rising Pune Supergiants': 'Rising Pune Supergiant'  # Spelling mistake
}
df['team1'] = df['team1'].replace(team_mapping)
df['team2'] = df['team2'].replace(team_mapping)
df['winner'] = df['winner'].replace(team_mapping)   
```

```python
winning_percentage = ((total_wins / total_matches) * 100).round(2)
winning_percentage
```
