```python
df=pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Attendance']=df['Attendance'].fillna(df['Attendance'].median())
df['Attendance']=df['Attendance'].astype(int)
```

```python
df.isnull().sum()
df
```

```python
df.duplicated().sum()
```

```python
df["Participation"]=df["Participation"].map({
    "Low":0,
    "Medium":1,
    "High":2
})
```

```python
df["Extra Classes"]=df["Extra Classes"].map({
    "Yes":1,
    "No":0
})
```

```python
sns.scatterplot(
    x="Study Hours",
    y="Attendance",
    hue="Previous Marks",
    data=df,
    s=100
)
```

```python
x=df[["Study Hours", "Attendance","Previous Marks","Assignments Completed","Sleep Hours","Internet Usage (Hours)","Participation","Extra Classes"]]
y=df["Final Marks"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(x_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
prediction=model.predict(x_test)
prediction
```

```python
new_student = [[
    7,      # Study Hours
    88,     # Attendance
    78,     # Previous Marks
    7,      # Assignments Completed
    7,      # Sleep Hours
    3,      # Internet Usage (Hours)
    1,      # Participation (High = 1, अगर Encoding की है)
    1       # Extra Classes (Yes = 1)
]]
```

```python
predict=model.predict(new_student)
predict
```

```python
from sklearn.metrics import r2_score
```

```python
r2=r2_score(y_test, prediction)
r2
```
