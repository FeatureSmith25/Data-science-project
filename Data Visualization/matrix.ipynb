```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

Heat map

```python
flights=sns.load_dataset('flights')
flights
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
tipscorr=tips[['total_bill','tip','size']]
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(),annot=True)
```

```python
sns.clustermap(tipscorr.corr(),annot=True)
```

```python
pvtflight=flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
pvtflight
```

```python
sns.heatmap(pvtflight)
```
