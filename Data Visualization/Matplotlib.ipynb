```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x=np.linspace(0,5,11)
x
y=x**2
y
```

```python
plt.plot(x,y)
plt.title("hello")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

Subplots in matplotlib


```python
plt.subplot(2, 2, 1)
plt.plot(y,x)
plt.subplot(2, 2, 4)
plt.plot(x,y)
plt.subplot(2, 2, 2)
plt.plot(x**2,x**2)
plt.subplot(2, 2, 3)
plt.plot(y**2,y**2)
```

Object oriented way


```python
fig = plt.figure()
axis1=fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2=fig.add_axes([0.3,0.4,0.3,0.3])
axis2.plot(x,y)
```

Types of plot

```python
fig = plt.figure(figsize=(20,8),dpi=100)
axis1=fig.add_axes([0.1,0.1,0.4,0.8])
axis1.plot(x,y)
axis2=fig.add_axes([0.1,0.1,0.3,0.3])
axis2.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.hist(x)
```

```python
plt.boxplot(x)
```

Saving the plots

```python
fig1=plt.figure()
ax=fig1.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("Basic_plot.png")
```
