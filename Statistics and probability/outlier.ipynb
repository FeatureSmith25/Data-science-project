```python
import numpy as np
import seaborn as sns
import pandas as pd
arr=np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
arr
```

```python
Q1=np.percentile(arr,25)
Q3=np.percentile(arr,75)
```

```python
Q1,Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
UF=Q3+1.5*IQR
LF=Q1-1.5*IQR
```

```python
UF,LF
```

```python
l=[]
for i in arr:
    if i>=LF and i<=UF:
        l.append(i)
```

```python
l
```

```python
arr1=np.array(l)
```

```python
arr,arr1
```

```python
sns.boxenplot(x=arr)
```

```python
sns.boxenplot(x=arr1)
```
