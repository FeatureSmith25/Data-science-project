```python
import numpy as np
from scipy.stats import norm
```

```python
u=150
x=145 
n=10
s=8
```

```python
T_value=(x-u)/(s/np.sqrt(n))
print(T_value)
```

```python
p_value=2*(1-norm.cdf(abs(T_value)))
p_value
```

```python
if p_value<T_value:
    print("I will reject the null hypothesis")
else:
    print("I will Accept the null hypothesis")
```
