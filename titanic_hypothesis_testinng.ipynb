```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import scipy.stats as stats
```

```python
df=sns.load_dataset('titanic')
print(df.head())
```

```python
df['age']=df['age'].fillna(df['age'].median(),inplace=True)
df['deck']=df['deck'].fillna(df['deck'].mode()[0],inplace=True)
df['embarked']=df['embarked'].fillna(df['embarked'].mode()[0],inplace=True)
df['embark_town']=df['embark_town'].fillna(df['embark_town'].mode()[0],inplace=True)
```

```python
female=df[df['sex']=='female']['survived']
male=df[df['sex']=='male']['survived']
```

```python
t_stat, p_value=stats.ttest_ind(female, male)
```

```python
if p_value < 0.05:
    print("We reject the null hypothesis. There is a significant difference in survival rates between males and females.")
else:
    print("We fail to reject the null hypothesis. There is no significant difference in survival rates between males and females.")
```
